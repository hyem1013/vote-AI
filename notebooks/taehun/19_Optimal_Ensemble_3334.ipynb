```python
import pandas as pd
import numpy as np
import os

# 1. 경로 설정
DATA_PATH = "../../data/raw/"
SUB_PATH = "./submissions/"

# 2. 앙상블 대상 모델 로드 (최종 검증된 3개 모델)
try:
    # m1: Exp11-SOTA (0.78116)
    # m2: Exp12-AUC (0.78108)
    # m4: Exp15-SNN (0.77077)
    m1 = pd.read_csv(f"{SUB_PATH}0130-1923.csv")['voted']
    m2 = pd.read_csv(f"{SUB_PATH}Pure_0781_AUC_0.77322.csv")['voted']
    m4 = pd.read_csv(f"{SUB_PATH}Diversity_Focal_SNN_AUC_0.7708.csv")['voted']
    print("최적화 대상 모델 로드 완료.")
except FileNotFoundError as e:
    print(f"파일을 찾을 수 없습니다: {e}")

# 3. 수학적 최적 가중치 적용 (33:33:34)
final_voted = (m1 * 0.3300) + (m2 * 0.3299) + (m4 * 0.3400)

# 4. 제출 파일 생성
submission = pd.read_csv(f"{DATA_PATH}sample_submission.csv")
submission['voted'] = final_voted

# 파일명 규칙 준수: [번호]_[유형]_[가중치]
output_filename = f"{SUB_PATH}19_Optimal_Ensemble_3334_.csv"
submission.to_csv(output_filename, index=False)

print("-" * 50)
print(f"메인 리더보드 저격용 최종 파일 생성 완료!")
print(f"파일명: {output_filename}")
print(f"최종 비율: m1(33%) : m2(32.99%) : m4(34%)")
print(f"예상 로컬 AUC: 0.77351")
print("-" * 50)
```
